```python
from tqdm import tqdm
```

```python
for p in tqdm(glob(in_path+'/*')):
    p_name = os.path.basename(os.path.normpath(p))
#     print(p_name)
    no_folders = int(len(glob(p+'/*'))/74)
    for i in range(no_folders):
        out_path_name = os.path.join(out_path,p_name+'_'+str(i))
        os.mkdir(out_path_name)
        for j, file in enumerate(glob(p+'/*')):
#             print(j)
            if j == 74:
                j = 0
                break
            shutil.move(file,out_path_name)
```
